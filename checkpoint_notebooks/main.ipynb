```python
import streamlit as st

# Sample data for recommendations based on tags
recommendations = {
    "Vegetarian": ["Vegetable Stir Fry", "Avocado Toast", "Vegetarian Pizza"],
    "Vegan": ["Vegan Salad", "Chickpea Curry", "Tofu Stir Fry"],
    "Gluten-Free": ["Grilled Chicken Salad", "Quinoa Bowl", "Zucchini Noodles"],
    "Low-Carb": ["Keto Burger", "Cauliflower Rice", "Baked Salmon"],
    "High-Protein": ["Grilled Chicken", "Protein Shake", "Egg Muffins"]
}

# Streamlit layout
st.title("Dish Recommender System")
st.write("Check your preferences to see personalized recommendations.")

# Layout for preferences and recommendations
selected_preferences = []
for tag in recommendations.keys():
    if st.checkbox(tag):
        selected_preferences.append(tag)

# Render separate rows for each tag's recommendations
for tag in selected_preferences:
    st.subheader(f"Recommendations for {tag}:")
    st.write(", ".join(recommendations[tag]))

# Final combined recommendations based on all selected tags
if selected_preferences:
    st.subheader("Final Combined Recommendations:")
    combined_recommendations = set()
    for tag in selected_preferences:
        combined_recommendations.update(recommendations[tag])
    st.write(", ".join(combined_recommendations))
else:
    st.write("Select preferences to get recommendations.")

# Display layout similar to Netflix-style with horizontal scrolling
st.markdown("---")
st.header("Netflix-Style Layout")
for tag, dishes in recommendations.items():
    st.subheader(tag)
    cols = st.columns(len(dishes))
    for col, dish in zip(cols, dishes):
        with col:
            st.image("https://via.placeholder.com/150", use_column_width=True)  # Replace with your dish images
            st.write(dish)
```
